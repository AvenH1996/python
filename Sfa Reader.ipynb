```python
import pdfplumber
import pandas as pd
import os
import re
fileList = os.listdir()
fileList = [i for i in os.listdir() if i[-3:] == 'pdf'or i[-3:] == 'PDF']

for i in fileList:
    with pdfplumber.open(i) as pdf:
        page_count = len(pdf.pages)
        f = open(i.replace('PDF','txt'), 'w')
        for page in pdf.pages:
            results = page.extract_words()
            try:
                for  word in results:
                    print(word['text']+'\n')
            except  UnicodeEncodeError:
                print(results)


            # print(pdf_table)
        f.close()
```
